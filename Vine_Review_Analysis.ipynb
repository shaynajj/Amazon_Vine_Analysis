```python
# Dependencies and Setup
import pandas as pd
import os
import numpy as np

# File to Load
vine_data_to_load = "Resources/vine_table.csv"

# Read the Data and store into a Pandas DataFrame
vine_data_df = pd.read_csv(vine_data_to_load)
vine_data_df
```

```python
total_votes_df = vine_data_df[vine_data_df["total_votes"] >= 20]
total_votes_df
```

```python
helpful_votes_df = total_votes_df[(total_votes_df["helpful_votes"]/total_votes_df["total_votes"]) >= .5]
helpful_votes_df
```

```python
vine_program_df = helpful_votes_df[helpful_votes_df["vine"] == 'Y']
vine_program_df
```

```python
non_vine_program_df = helpful_votes_df[helpful_votes_df["vine"] == 'N']
non_vine_program_df
```

```python
total_reviews = helpful_votes_df["review_id"].count()
total_reviews
```

```python
five_star_vine_program_df = vine_program_df[vine_program_df["star_rating"] == 5.0]
five_star_vine_program_df
```

```python
five_star_vine_reviews = five_star_vine_program_df["review_id"].count()
five_star_vine_reviews
```

```python
five_star_non_vine_program_df = non_vine_program_df[non_vine_program_df["star_rating"] == 5.0]
five_star_non_vine_program_df
```

```python
five_star_non_vine_reviews = five_star_non_vine_program_df["review_id"].count()
five_star_non_vine_reviews
```

```python
five_star_vine_percent = five_star_vine_reviews / total_reviews
five_star_vine_percent
```

```python
five_star_non_vine_percent = five_star_non_vine_reviews / total_reviews
five_star_non_vine_percent
```
